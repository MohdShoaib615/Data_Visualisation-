```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import plotly.express as px
```

# ***HR_DATA_SET***

# ***VISUALISATION***

```python
df=pd.read_csv('/content/HR_Data_set_cleaned')
df
```

```python
# VISUAL-1

# Visualising the salary-structure for termination reason of employees

plt.bar(x=df['Termd'],height=df['Salary'])
```

```python
# Visualisation-2 

# Visualising the Empsatisfaction of the employees
sns.boxplot(x=df['Termd'],y=df['EmpSatisfaction'])
```

```python
# Visualisation-3

# Visualising the most popular department

sns.histplot(x=df['Salary'])
```

```python
# Visualisation-4

# Visualising the Reason of termination of employees in each department with respect to their salary
plt.rcParams['figure.figsize']=(10,10)

sns.boxplot(x=df['Department'],y=df['Salary'],hue=df['Termd'])
```

```python
# Visualisation-5

# Visualising the Reason of termination of employees in each Position with respect to their salary
px.scatter(df,x='Position',y='Salary',color='Termd')
```

```python
# Visualisation-6

# Visualising the Reason for termination of employees with respect to Salary for Males & Females
px.box(x=df['Sex'],y=df['Salary'],color=df['Termd'])
```

```python
# VISUALISATION-7
# Visualising the ProjectsCount as a reason for termination of employees
px.box(x=df['Termd'],y=df['SpecialProjectsCount'])
```

```python
# VISUALISATION-8
# Visualising the termination of employees in each Department as per their Absences

sns.barplot(x=df['Department'],y=df['Absences'],hue=df['Termd'])
```

```python
# VISUALASATION-9

# Visualising the Termination of employees for performance_score in terms of their satisfaction
px.bar(df, x="PerformanceScore", y="EmpSatisfaction", color="Termd",
             pattern_shape="PerformanceScore", pattern_shape_sequence=["+", "."])
```

```python
# VISUALISATION-10

# Visualising the Departments available

fig = px.pie(names=df['EmploymentStatus'], title='Total % of Departments')
fig.show()
```

# ***REMARKS FOR VISUALISATION***

---



```python
# Visualisation-1---By analysing the Salary column and the Employment status whether active/terminated, we concluded that the working employees are getting more salaries, that's why they are still active.
# Visualasation-2---By Visualising the Empsatisfaction of the employees, we concluded that those employees who are terminated have low satisfaction.
# Visualisation-3-- With this visualisation, we conclude that mostly all the employees have low salary, only few of them have high salary.
# Visualasation-4---With this visualisation, we concluded that the departments having high/outliers(high) salaries have the more active employees.
# Visualisation-5---With this visualisation, we concluded that there are many positions where no employee is being terminated or no employee is active in some of the positions.
# Visualasation-6---With this visualisation, we concluded that due to many outliers for Males & Females in terms of their salary, the other employees are leaving as their salary is comparatively low.
# Visualisation-7---With this visualisation, we concluded that due to many no of projects count, the employees are able to work, and those having less projects are being terminated.
# Visualasation-8---With this visualisation, we concluded that in mosly all the department, employees having less Absences are active and those having high absences are terminated
# Visualisation-9---With this visualisation, we concluded that emplyees who fully meets the performance_score are active
# Visualasation-10---With this visualisation, we concluded that inspite of all factors, a large fraction of employees are still active.
```

# ***AIRBNB_DATASET***

```python
df2=pd.read_csv('/content/Airbnb_cleaned')
df2
```

```python
# VISUALISATION-1

# Visualising the price of each room_type

plt.bar(x=df2['room_type'],height=df2['price'])
```

```python
# VISUALISATION-2

# Visualising the review_per_month based on room_type
 
px.histogram(x=df2['room_type'],y=df2['reviews_per_month'])
```

```python
# VISUALISATION-3

# Visualising the price for neighbourhood_group based on room_type

sns.boxplot(x=df2['neighbourhood_group'],y=df2['price'],hue=df2['room_type'])
```

```python
# VISUALISATION-4

# Visualising the counts based on minimum_nights
sns.displot(x=df2['minimum_nights'])
```

```python
# VISUALISATION-5

# Visualising the effect of minumum_nights on reviews  

px.histogram(x=df2['minimum_nights'],y=df2['number_of_reviews'])
```

```python
# VISUALISATION-6

# Visualising the price of with respect to each neighbour_group
plt.bar(x=df2['neighbourhood_group'],height=df2['price'])
```

```python
# VISUALISATION-7

# Visualising the reviews of neighbourhood_group for each room_type

px.box(df2,x='neighbourhood_group',y='number_of_reviews',color='room_type')
```

```python
# VISUALISATION-8

# Visualising the reviews_per_month based on price

plt.bar(x=df2['price'],height=df2['reviews_per_month'])
```

```python
# VISUALISATION-9

# Visualising the price based on number_of_reviews for each room_type
px.scatter(df2,x='number_of_reviews',y='price',color='room_type',marginal_x = 'box', marginal_y = 'violin')
```

```python
# VISUALISATION-10

# Visualising the counts of rooms based on their availability_365

px.histogram(df2,x='availability_365')
```
